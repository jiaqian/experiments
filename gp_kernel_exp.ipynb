```python
import numpy as np
from scipy.spatial import distance
import matplotlib.pyplot as plt
```

```python
# Define the kernel
def kernel(xa, xb, the0, the1, the2, the3):
    # L2 distance (Squared Euclidian)
    ker = the0 * np.exp(-0.5 * the1 *distance.cdist(xa, xb, 'sqeuclidean')) + the2 + the3* xa * xb
    return ker
```

```python
nb_of_samples = 100  
number_of_functions = 5  # Number of functions to sample
the0, the1, the2, the3 = 1, 64, 0 ,0
# Independent variable samples
X = np.expand_dims(np.linspace(-1, 1, nb_of_samples), 1)
sigma = kernel(X, X, the0, the1, the2, the3)  # Kernel of data points
```

```python
sigma.shape
```

```python
ys = np.random.multivariate_normal(
    mean=np.zeros(nb_of_samples), cov=sigma, 
    size=number_of_functions)
```

```python
plt.plot(X,ys[0,:])
plt.plot(X,ys[1,:])
plt.plot(X,ys[2,:])
plt.plot(X,ys[3,:])
plt.plot(X,ys[4,:])
```
